```python
# secret share a list of values among two servers
def share_list(xs):
    # secret share each element of the list
    all_shares = [share(secret=x, n=2) for x in xs]

    # make two lists of shares: one for server 1, and the other for server 2
    shares1, shares2 = zip(*all_shares)

    # return the shares
    return shares1, shares2

share_list([1,2,3])
```

```python
def sum_shares(shares):
    # sum the shares in the list of shares
    return gf(shares).sum()

# share the list values among two servers
shares1, shares2 = share_list([1,2,3])

# server 1 calculates sum1
sum1 = sum_shares(shares1)

# server 2 calculates sum2
sum2 = sum_shares(shares2)

# client reconstructs total sum from the results
reconstruct([sum1, sum2])
```
